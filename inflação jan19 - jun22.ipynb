```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('tabela1737.xlsx')

df.head()
```

```python
month = df['Month']
ipca = df['IPCA']
month.head(10)
```

```python
plt.plot(month, ipca, label = 'IPCA')
plt.style.use("ggplot")
plt.xticks(rotation=45)
plt.xlabel('Fonte: SIDRA IBGE')
plt.yscale('linear')
plt.title('IPCA - Variação acumulada em 12 meses\n (jan/2019 - jun/2022)')
plt.ylabel('%')
plt.axhline(3.5, color = 'red', linestyle = '--', label ='Meta')
plt.legend()
```
